```python
Tn = np.arange(0.75, 1.05, 0.05)
Vn = np.linspace(1/3+1e-1, 5, 2000)
Pn_list = [(8*Ti)/(3*Vn-1)-3/(Vn**2) for Ti in Tn]

plt.figure(dpi=150)
for i,pn in enumerate(Pn_list):
  plt.plot(Vn, pn, label="$T/T_c=$"+str(round(Tn[i],2)))
plt.xlabel(r'$V/V_c$')
plt.ylabel(r'$P/P_c$')
plt.ylim([-1,3])
#plt.title(r'$\hat{P] vs \hat{V}$')
font_props = FontProperties(style='italic', family='Times New Roman')
plt.legend(prop=font_props)
plt.legend()
```

## Examples of Graphs for Ar and H2O

```python
a_argon = 0.136       # Pam^6mol^-2
b_argon = 0.03219e-3  # m^3mol^-1
Tc_argon = 150.85     # K
a_water = 0.5536      # Pam^6mol^-2
b_water = 0.03049e-3  # m^3mol^-1
Tc_water = 647.1      # K

# Constants
R = 8.314    # J/(molK)
n = 1.0      # mol

# Volume range
V_water = np.linspace(b_water*n+b_water*n*0.05, 0.003, 1000) # m^3
V_argon = np.linspace(b_argon*n+b_argon*n*0.05, 0.002, 1000) # m^3


# Select two temperatures below Tc, one at Tc, and two above Tc for each gas
T_values_argon = [Tc_argon*0.6, Tc_argon*0.8, Tc_argon, Tc_argon*1.2, Tc_argon*1.4]
T_values_water = [Tc_water*0.6, Tc_water*0.8, Tc_water, Tc_water*1.2, Tc_water*1.4]

# Van der Waals equation for pressure
def pressure(V, a, b, T):
    return (n*R*T) / (V-n*b) - a*(n**2/V**2)

# Create the plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot for Argon
for T in T_values_argon:
    P_argon = pressure(V_argon, a_argon, b_argon, T)
    axs[0].plot(V_argon, P_argon, label=f"T = {T-273.15:.1f} C")
axs[0].set_xlabel('Volume (m^3)')
axs[0].set_ylabel('Pressure (Pa)')
axs[0].set_title('P-V Diagram for Argon using the Van der Waals Model')
axs[0].legend()
axs[0].grid(True)
axs[0].set_ylim([-1e7,4e7])
axs[0].set_xlim([0,4e-4])

# Plot for Water
for T in T_values_water:
    P_water = pressure(V_water, a_water, b_water, T)
    axs[1].plot(V_water, P_water, label=f"T = {T-273.15:.1f} C")
axs[1].set_xlabel('Volume (m^3)')
axs[1].set_ylabel('Pressure (Pa)')
axs[1].set_title('P-V Diagram for Water using the Van der Waals Model')
axs[1].legend()
axs[1].grid(True)
axs[1].set_ylim([-1e7,4e7])
axs[1].set_xlim([0,1e-3])

plt.tight_layout()
plt.show()
```

## Schroeder Simulation

```python
# Read the new .txt file with correct headers and separator
data3 = pd.read_csv("sim_schroeder3.txt", sep="\t", skiprows=1, names=["time", "N", "V", "E", "T", "P", "KE", "PE", "GE", "px", "py"])

# Calculate average temperature and its standard deviation
average_T = data3['T'].mean()
std_dev_T = data3['T'].std()

print(f"Average Temperature: {average_T}")
print(f"Temperature Standard Deviation: {std_dev_T}")

# Create a Pressure vs Volume plot
plt.figure(figsize=(10,5))
plt.plot(data3['V'], data3['P'])
plt.title('Pressure vs Volume')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.grid(True)
plt.show()

# Create a Total Energy plot with no x-axis
plt.figure(figsize=(10,5))
plt.plot(data3['E'])
plt.title('Total Energy')
plt.xlabel('Index')
plt.ylabel('Total Energy')
plt.grid(True)
plt.show()

# Reverse the order of the energy data
data3['E_reversed'] = data3['E'].values[::-1]

# Create a Total Energy plot with "Steps" as x-axis
plt.figure(figsize=(10,5))
plt.plot(data3['E_reversed'])
plt.title('Total Energy')
plt.xlabel('Steps')
plt.ylabel('Total Energy')
plt.grid(True)
plt.show()
```
